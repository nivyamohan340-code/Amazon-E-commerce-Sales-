#### **Skewness**
Statistical measure that tells how symmetric or asymmetric the distribution of a numerical variable is.

```python
# Check skewness,
print("\nSkewness:")
print(df[numeric_cols].skew())
```

#### **Types of Skewness**


 **1.Zero Skewness (Symmetric Distribution)**

Data is evenly distributed on both sides of the mean.

Mean ≈ Median.

Looks like a normal bell curve.

**2.Positive Skewness (Right-Skewed)**

Long tail on the right side.

Few very large values.

Mean > Median.

Common in sales data (few very high-value orders).

**3.Negative Skewness (Left-Skewed)**

Long tail on the left side.

Few very small values.

Mean < Median.

#### **Why Skewness is Important in Amazon Sales Project**

Helps identify uneven sales distribution.

Reveals presence of high-value outlier orders.

Indicates whether data transformation may be needed before modeling.

## **Outlier Check (IQR Method)**

```python
Q1 = df['Total_Sales'].quantile(0.25)
Q3 = df['Total_Sales'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['Total_Sales'] < Q1 - 1.5*IQR) | (df['Total_Sales'] > Q3 + 1.5*IQR)]
print("Number of outliers in Total_Sales:", len(outliers))
```

## **Multivariate Analysis**

Multivariate analysis studies relationships among three or more variables simultaneously to identify combined effects and deeper patterns.

For the Amazon E-commerce Sales dataset, multivariate analysis was carried out to understand how sales performance is influenced by multiple factors such as **product category**, **region**, **promotion usage**, **fulfilment type**, and **time period**.

### **1.Category × Region × Total Sales**

```python
cat_region_sales = df.pivot_table(
    values='Total_Sales',
    index='Category',
    columns='Region',
    aggfunc='sum',
    fill_value=0
)
print(cat_region_sales)
```

### **2.Month × Category × Total Sales**

```python
month_category_sales = df.pivot_table(
    values='Total_Sales',
    index='Order_Month',
    columns='Category',
    aggfunc='sum',
    fill_value=0
)
print(month_category_sales)
```

### **3.Promotion × Category × Average Sales**

```python
promo_category_sales = df.pivot_table(
    values='Total_Sales',
    index='Category',
    columns='Promo_Used',
    aggfunc='mean',
    fill_value=0
)
print(promo_category_sales)
```

Multivariate analysis was performed using pivot tables to examine the combined impact of product category, region, promotion usage, fulfilment type, and order month on total sales. The analysis helped identify region-wise top-selling categories, seasonal variations in product demand, effectiveness of promotional strategies, and fulfilment performance across different order statuses. This provided deeper insights into key factors influencing overall revenue.

## **Visualization**

### **1. Monthly Sales Trend(Line chart)**

The linechart illustrates the trend of sales over the months,highlighting patterns,fluctuations and overall growth.
Each point on the line represents the total sales for a particular month,allowing us to observe peak periods,seasonal trends,and potential dips in performance.
This visualization helps in identifying consistent growth, periods of high demand, and months that may require targeted strategies to boost sales.

```python
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Amount'].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(color='green')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales Amount")
plt.show()
```

The line chart represents the monthly sales trend over time. Sales show a sharp increase from March to April, followed by a slight decline in May and June, indicating peak sales in April and a minor drop afterward.

### **2. Top 10 Categories by Sales (Bar Plot)**

The bar plot displays the top ten product categories ranked by total sales. Each bar represents a category, with its height indicating the total sales value. This visualization makes it easy to compare category performance, identify the highest revenue-generating categories, and understand which product segments contribute most to overall sales. Such insights help in inventory planning, marketing focus, and strategic decision-making.

```python
top_cat = df.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_cat.index, y=top_cat.values)
plt.title("Top 10 Product Categories by Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45)
plt.show()
```

The bar chart shows the top 10 product categories by total sales. Set and Kurta generate the highest sales, indicating strong customer demand, while the remaining categories contribute comparatively lower sales. Overall, sales are concentrated mainly in the top three categories.

### **3. Order Status Distribution (Bar Plot)**

The bar plot represents the distribution of orders across different order statuses such as delivered, shipped, pending, or cancelled. Each bar shows the count of orders in a particular status category. This visualization helps in understanding order fulfillment performance, identifying delays or bottlenecks in processing, and evaluating the efficiency of the supply chain workflow.

```python
order_status_counts=df['Status'].value_counts()
plt.figure(figsize=(14,6))
order_status_counts.plot(kind='bar')
plt.xticks(rotation=45, ha='right')
plt.title("Order Status Distribution",fontweight='bold')
plt.xlabel("Order status",fontweight='bold')
plt.ylabel(" Number of Order",fontweight='bold')
plt.tight_layout()
plt.show()
```

The bar chart displays the distribution of order statuses. Most orders are successfully delivered, while a smaller portion are cancelled or returned, and very few remain pending or in transit, indicating efficient order fulfillment overall.

### **4. Fulfilment Method vs Revenue (Bar Plot)**

The bar plot compares total revenue generated through different fulfilment methods, such as merchant fulfilment and platform fulfilment. Each bar represents a fulfilment method, and its height indicates the total revenue earned. This visualization helps in identifying which fulfilment strategy contributes more to overall sales, enabling better operational planning and optimization of delivery processes.

```python
ful_sales = df.groupby('Fulfilment')['Amount'].sum()

plt.figure(figsize=(7,5))
sns.barplot(x=ful_sales.index, y=ful_sales.values)
plt.title("Revenue by Fulfilment Method")
plt.xlabel("Fulfilment Method")
plt.ylabel("Total Revenue")
plt.show()
```

The bar chart compares total revenue generated by different fulfilment methods. Amazon fulfilment contributes significantly higher revenue than Merchant fulfilment, indicating that Amazon-handled deliveries drive most of the sales

### **5. Quantity vs Sales Amount (Scatter Plot)**

The scatter plot visualizes the relationship between the quantity of products sold and the corresponding sales amount. Each point represents an order, where the x-axis shows the quantity purchased and the y-axis shows the sales value. This visualization helps in understanding how sales revenue changes with order quantity, identifying bulk purchase patterns, and spotting outliers such as unusually high-value orders. It also provides insight into whether higher quantities consistently lead to increased sales revenue.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Qty', y='Amount', data=df)
plt.title("Quantity vs Sales Amount")
plt.xlabel("Quantity Ordered")
plt.ylabel("Sales Amount")
plt.show()
```

The scatter plot shows the relationship between quantity ordered and sales amount. Higher quantities generally correspond to higher sales amounts, with a few high-quantity orders generating exceptionally large sales, indicating a positive correlation between quantity and revenue.

### **6. Category vs Fulfilment Heatmap (Pivot + Heatmap)**

The heatmap visualizes the relationship between product categories and fulfilment methods based on total sales revenue. Using a pivot table, categories are plotted along one axis and fulfilment methods along the other, with color intensity representing the sales amount. Darker shades indicate higher revenue contribution. This visualization helps in identifying which fulfilment methods perform best for specific product categories, revealing operational strengths and opportunities for optimization in the supply chain process.

```python
pivot = pd.pivot_table(df,
                       values='Amount',
                       index='Category',
                       columns='Fulfilment',
                       aggfunc='sum',
                       fill_value=0)

plt.figure(figsize=(10,6))
sns.heatmap(pivot, cmap='YlGnBu')
plt.title("Category vs Fulfilment – Sales Heatmap")
plt.xlabel("Fulfilment Method")
plt.ylabel("Category")
plt.show()
```

The heatmap shows sales intensity across product categories and fulfilment methods. Set and Kurta categories under Amazon fulfilment generate the highest sales, while most other categories and merchant fulfilment show lower sales, indicating Amazon fulfilment drives stronger performance across major categories.

### **7. Correlation Heatmap (Multivariate)**

The correlation heatmap represents the strength and direction of relationships between multiple numerical variables in the dataset. Each cell displays a correlation coefficient, with color intensity indicating whether the relationship is positive, negative, or neutral. Strong positive correlations appear in darker warm shades, while strong negative correlations appear in cooler shades. This visualization helps in identifying highly related features, detecting multicollinearity, and understanding how different variables influence sales performance. It provides valuable insights for feature selection and strategic decision-making.

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Qty','Amount']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Quantity and Sales Amount")
plt.show()
```

The correlation heatmap shows a very weak positive correlation (≈ 0.044) between quantity ordered and sales amount. This indicates that sales value is not strongly dependent on the number of items ordered, suggesting that product price has a greater impact on total sales than quantity.

### **11. State-wise Sales Distribution (Box Plot)**

The box plot visualizes the distribution of sales amounts across different states. Each box represents a state and shows the median, quartiles, and range of sales values, while outliers indicate unusually high or low sales transactions. This visualization helps in understanding sales variability between states, identifying states with consistently high or low sales, and detecting irregular sales patterns. It provides valuable insights for regional performance evaluation and targeted business strategies.
